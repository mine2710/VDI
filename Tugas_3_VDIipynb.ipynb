```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer
from matplotlib import pyplot as plt
```

```python

df = pd.read_excel('/content/dataset.xlsx', engine='openpyxl')

df.head()
```

```python
sns.lineplot(x='Tanggal', y='Penjualan (pcs)', data=df)
plt.title('Tren Penjualan')
plt.xlabel('Tanggal')
plt.ylabel('Penjualan')
plt.show()
```

```python
# Line chart untuk tren penjualan
sns.lineplot(x='Tanggal', y='Penjualan (pcs)', data=df)
plt.title('Tren Penjualan (pcs)')
plt.show()

sns.barplot(x='Tanggal', y='Penjualan (pcs)', data=df)
plt.title('Tren Penjualan (pcs)')
plt.show()

# Histogram untuk distribusi Penjualan (pcs)
sns.histplot(df['Penjualan (pcs)'], bins=20, kde=True)
plt.title('Distribusi Penjualan (pcs)')
plt.show()

# Bar chart untuk membandingkan Penjualan (pcs) berdasarkan hari
sns.barplot(x='Hari', y='Penjualan (pcs)', data=df)
plt.title('Penjualan (pcs) per Hari')
plt.show()

# Scatter plot untuk melihat hubungan antara curah hujan dan Penjualan (pcs)
sns.scatterplot(x='Curah Hujan (mm)', y='Penjualan (pcs)', data=df)
plt.title('Hubungan Curah Hujan dan Penjualan (pcs)')
plt.show()
```

```python
sns.barplot(x='Hari', y='Penjualan (pcs)', data=df)
plt.title('Penjualan (pcs) per Hari')
plt.show()

sns.histplot(df['Penjualan (pcs)'], bins=20, kde=True)
plt.title('Distribusi Penjualan (pcs)')
plt.show()

sns.lineplot(x='Tanggal', y='Penjualan (pcs)', data=df)
plt.title('Tren Penjualan (pcs)')
plt.show()
```

```python
sns.scatterplot(x='Curah Hujan (mm)', y='Penjualan (pcs)', data=df)
plt.title('Hubungan Curah Hujan dan Penjualan (pcs)')
plt.show()

sns.barplot(x='Hari', y='Penjualan (pcs)', data=df)
plt.title('Penjualan (pcs) per Hari')
plt.show ()

sns.histplot(df['Penjualan (pcs)'], bins=20, kde=True)
plt.title('Distribusi Penjualan (pcs)')
plt.show ()
```

```python
sns.pairplot(df, hue='Kegiatan')
plt.show()
```
